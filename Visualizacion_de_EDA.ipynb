```
#Llamamos al set de datos publicos
zoo_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'
zoo_name = ['animal name',
            'hair',
            'feathers',
            'eggs',
            'milk',
            'airborne',
            'aquatic',
            'predator',
            'toothed',
            'backbone',
            'breathes',
            'venomous',
            'fins',
            'legs',
            'tail',
            'domestic',
            'catsize',
            'type']
```

```
import pandas as pd
```

```
df_zoo = pd.read_csv(zoo_data)
```

```
df_zoo.head(10)
```

```
df_zoo.columns = zoo_name
```

```
df_zoo.head(5)
```

```
import altair as alt
```

```
plot01 = alt.Chart(df_zoo).mark_line().encode(
    x = 'animal name',
    y = 'legs'
)
```

```
plot01
```

```
#Grafica de barras
plot02 = alt.Chart(df_zoo).mark_bar().encode(
    x = 'animal name',
    y = 'legs'
)
```

```
plot02
```

```
import plotly.express as px
```

```
px.pie(df_zoo,values='aquatic', names='animal name', title='Grafica de pie')
```
